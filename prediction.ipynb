```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import load_model
```

```python
max_len = 200
```

```python
word_index = imdb.get_word_index()
reverse_word_index = {value: key for key, value in word_index.items()}
```

```python
model = load_model('model_rnn.h5')
model.summary()
```

Preparing the data to fed into trained Model

```python
def decode_review(encoded_review):
    return ' '.join([reverse_word_index.get(i - 3, '?') for i in encoded_review])


def preprocess_text(text):
    words = text.lower().split()
    encoded_review = [word_index.get(word, 2) + 3 for word in words]
    padded_review = pad_sequences([encoded_review], maxlen=max_len,padding='pre')
    return padded_review
```

```python
def predict_sentiment(review):
    preprocessed_input=preprocess_text(review)

    prediction=model.predict(preprocessed_input)

    sentiment = 'Positive' if prediction[0][0] > 0.5 else 'Negative'
    
    return sentiment, prediction[0][0]
```

```python
review = "An absolute triumph of storytelling. It balances hilarious comedy with deep, heartfelt drama flawlessly. A rare, perfect five-star cinematic experience."

sentiment,score=predict_sentiment(review)

print(f'Review: {review}')
print(f'Sentiment: {sentiment}')
print(f'Prediction Score: {score}')
```
